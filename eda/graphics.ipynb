```python
import pandas as pd
ps = pd.read_csv('df_after_fe.csv')
```

```python
ps['hidden_work_index'] = (
    ps['passes_key'] * 0.3 +
    ps['duels_won'] * 0.2 +
    (ps['dribbles_success'] / (ps['dribbles_attempts'] + 1)) * 0.3 +
    (1 / (ps['fouls'] + 1)) * 0.2
)

rating_by_position = ps.groupby("position")['player_rating'].mean()
print("Средний рейтинг по позициям:")
print(rating_by_position.sort_values(ascending=False))

hidden_work_by_position = ps.groupby("position")['hidden_work_index'].mean()
print("\nНевидимая работа по позициям:")
print(hidden_work_by_position.sort_values(ascending=False))
```

```python
comparison = pd.DataFrame({
    'Средний рейтинг': rating_by_position,
    'Невидимая работа': hidden_work_by_position
}).round(2)

comparison = comparison.sort_values('Невидимая работа', ascending=False)
```

```python
plt.figure(figsize=(12, 6))

x = range(len(comparison))
width = 0.35

plt.bar([i - width/2 for i in x], comparison['Средний рейтинг'],
        width, label='Средний рейтинг', color='skyblue', alpha=0.8)
plt.bar([i + width/2 for i in x], comparison['Невидимая работа'],
        width, label='Невидимая работа', color='lightcoral', alpha=0.8)

plt.xlabel('Позиции')
plt.ylabel('Значение')
plt.title('Сравнение рейтинга и невидимой работы по позициям')
plt.xticks(x, comparison.index, rotation=45)
plt.legend()
plt.grid(True, alpha=1)
plt.tight_layout()
plt.show()
```

```python
positions = comparison.index
fig = go.Figure(
    data=[
        go.Bar(
            name='Средний рейтинг',
            x=positions,
            y=comparison['Средний рейтинг'],
            marker_color=BRAND_COLORS["white"]
        ),
        go.Bar(
            name='Невидимая работа',
            x=positions,
            y=comparison['Невидимая работа'],
            marker_color=BRAND_COLORS["g_main"]
        )
    ]
)

fig.update_layout(
    title_text='Сравнение рейтинга и невидимой работы по позициям',
    xaxis_title='Позиции',
    yaxis_title='Значение',
    barmode='group'
)

fig.show()
```
