```python
# определяем числовые, булевы и строковые колонки
num_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
bool_cols = df.select_dtypes(include=['bool']).columns.tolist()
obj_cols = df.select_dtypes(include=['object']).columns.tolist()
dt_cols  = df.select_dtypes(include=['datetime64[ns]']).columns.tolist()

# исключаем player_id из списка
for col_list in [num_cols, bool_cols, obj_cols, dt_cols]:
    if 'player_id' in col_list:
        col_list.remove('player_id')

# словарь правил агрегации
agg_rules = {}

# числовые — максимум
for c in num_cols:
    agg_rules[c] = 'max'

# булевые — максимум (True побеждает)
for c in bool_cols:
    agg_rules[c] = 'max'

# строки — первое ненулевое значение
for c in obj_cols:
    agg_rules[c] = lambda x: x.dropna().iloc[0] if x.dropna().shape[0] else np.nan

# даты — максимум (последняя известная)
for c in dt_cols:
    agg_rules[c] = 'max'

# собираем мастер-таблицу
player_master = df.groupby('player_id', as_index=False).agg(agg_rules)

print(f"✅ Сформирован словарь игроков: {len(player_master)} записей (из {df['player_id'].nunique()})")
```

```python
df = df.merge(player_master, on='player_id', suffixes=('', '_fill'))

for col in player_master.columns:
    if col != 'player_id' and col in df.columns:
        df[col] = df[col].fillna(df[f'{col}_fill'])
        df.drop(columns=[f'{col}_fill'], inplace=True, errors='ignore')

print("✅ Пропуски заполнены по мастер-словарю игроков.")
```

```python
# создаём копию, чтобы не потерять оригинальные значения
df['yellow_red_cards_filled'] = df['yellow_red_cards']

# если у игрока нет ни жёлтых, ни красных - точно нет жёлто-красных
mask_no_cards = (df['yellow_cards'] == 0) & (df['red_cards'] == 0)
df.loc[mask_no_cards, 'yellow_red_cards_filled'] = 0

# если есть хоть одна жёлтая или красная, но значение NaN — ставим 0
mask_some_cards = ((df['yellow_cards'] > 0) | (df['red_cards'] > 0)) & (df['yellow_red_cards'].isna())
df.loc[mask_some_cards, 'yellow_red_cards_filled'] = 0

# если есть жёлто-красная, а жёлтая или красная == 0 — исправляем на минимум 1
mask_conflict = (df['yellow_red_cards_filled'] > 0) & ((df['yellow_cards'] == 0) | (df['red_cards'] == 0))
df.loc[mask_conflict, ['yellow_cards', 'red_cards']] = df.loc[mask_conflict, ['yellow_cards', 'red_cards']].fillna(1)

print(f"yellow_red_cards заполнено у {df['yellow_red_cards_filled'].notna().sum()} из {len(df)} игроков")
```

```python
# числовые колонки, которые нужно привести к float
num_cols = [
    'player_height', 'player_weight',
    'shots_on', 'shots_total', 'goals', 'assists',
    'passes_total', 'passes_key',
    'duels_total', 'duels_won',
    'dribbles_attempts', 'dribbles_success',
    'fouls', 'yellow_cards', 'yellow_red_cards', 'red_cards',
    'sidelined_times', 'trophies_count',
    'years_remaining', 'age',
    'player_id', 'team_id', 'season'
]

for col in num_cols:
    if col in df.columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace('[^0-9.\-]', '', regex=True)  # оставляем и точку
            .replace('', np.nan)
            .astype(float)
        )


if 'player_rating' in df.columns:
    df['player_rating'] = pd.to_numeric(df['player_rating'], errors='coerce').astype(float)

bool_cols = ['injured', 'is_captain']
for col in bool_cols:
    if col in df.columns:
        df[col] = df[col].astype(bool)

date_cols = ['signed', 'expiration']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

print("типы колонок после нормализации:")
df.info()
```

```python
has_height = df['player_height'].notna()
has_weight = df['player_weight'].notna()

only_height = (has_height & ~has_weight).sum()
only_weight = (~has_height & has_weight).sum()
none_both   = (~has_height & ~has_weight).sum()
both_exist  = (has_height & has_weight).sum()

print(f"Есть рост, но нет веса: {only_height}")
print(f"Есть вес, но нет роста: {only_weight}")
print(f"Нет ни роста, ни веса: {none_both}")
print(f"Есть и рост, и вес: {both_exist}")

print(f"\nПроверка: сумма категорий = {only_height + only_weight + none_both + both_exist} / {len(df)}")
```

```python
mask_no_weight = df['player_weight'].isna()
mask_no_height = df['player_height'].isna()

weight_by_height = (
    df.dropna(subset=['player_height', 'player_weight'])
    .groupby('player_height')['player_weight']
    .apply(list)
    .to_dict()
)

mean_h, std_h = df['player_height'].dropna().mean(), df['player_height'].dropna().std()
mean_w, std_w = df['player_weight'].dropna().mean(), df['player_weight'].dropna().std()

def fill_weight(row):
    if pd.notna(row['player_weight']):
        return row['player_weight']
    h = row['player_height']
    # если рост известен и есть статистика
    if pd.notna(h) and h in weight_by_height and len(weight_by_height[h]) > 2:
        return int(np.random.choice(weight_by_height[h]))
    # если рост известен, но нет данных по этому росту
    elif pd.notna(h):
        return int(np.random.normal(mean_w, std_w))
    # если роста вообще нет — оба из глобального распределения
    else:
        h_new = int(np.random.normal(mean_h, std_h))
        w_new = int(np.random.normal(mean_w, std_w))
        df.at[row.name, 'player_height'] = h_new
        return w_new

# применяем
df['player_weight'] = df.apply(fill_weight, axis=1).round().astype('Int64')
df['player_height'] = df['player_height'].round().astype('Int64')

print(f"Пропуски роста и веса полностью заполнены: осталось {df['player_weight'].isna().sum()} весов и {df['player_height'].isna().sum()} ростов")
```

```python
df_work = df_backup.copy(deep=True)

# числовые столбцы и NA
num_cols = ['shots_total','shots_on','passes_total','passes_key','goals','assists','age','player_height','player_weight']
for c in num_cols:
    if c in df_work.columns:
        df_work[c] = pd.to_numeric(df_work[c], errors='coerce').fillna(0)

# возрастная кривая (пик 26)
age = df_work['age'].clip(16, 40)
age_factor = np.exp(-((age - 26) ** 2) / (2 * (8 ** 2)))

# позиции
pos = df_work['position'].fillna('Unknown').str.upper()
position_multipliers = {
    'FW': 1.20, 'ST': 1.20, 'CF': 1.20, 'LW': 1.18, 'RW': 1.18,
    'AM': 1.15, 'CAM': 1.15, 'LM': 1.10, 'RM': 1.10,
    'CM': 1.08, 'DM': 1.05, 'CDM': 1.05,
    'CB': 1.00, 'RB': 0.98, 'LB': 0.98, 'LWB': 1.02, 'RWB': 1.02,
    'GK': 0.85
}
pos_factor = pos.map(position_multipliers).fillna(1.0)

# лиги (строки как в данных)
league = df_work['league']
league_multipliers = {
    'England Premier League': 1.85,
    'Spain La Liga':          1.70,
    'Germany Bundesliga':     1.38,
    'Italy Serie A':          1.36,
    'France Ligue 1':         1.34,

    'Netherlands Eredivisie': 1.00,
    'Portugal Primeira Liga': 1.00,
    'Austria Bundesliga':     0.90,
    'Denmark Superliga':      0.82,

    'Saudi-Arabia Pro League': 0.56,
    'USA Major League Soccer': 0.78,

    'England Championship':    0.74,
    'Germany 2. Bundesliga':   0.64,
    'France Ligue 2':          0.66,
    'Italy Serie B':           0.68,
    'England League One':      0.62,
    'England League Two':      0.58,

    'Brazil Serie A':         0.92,
    'Argentina Liga Profesional Argentina': 0.90,
    'Mexico Liga MX':         0.88,
    'Russia Premier League':  0.86,
    'Sweden Allsvenskan':     0.86,
    'Croatia HNL':            0.84,
    'Colombia Primera A':     0.82,
    'Belgium Challenger Pro League': 0.72,
}
league_factor = league.map(league_multipliers).fillna(1.0)

# оценка числа матчей (грубая)
estimated_matches = np.maximum(df_work['shots_total'] / 3.5, 1)

# метрики на матч
g_pm = df_work['goals'] / estimated_matches
a_pm = df_work['assists'] / estimated_matches

# точность ударов
shot_acc = np.where(df_work['shots_total'] > 0,
                    df_work['shots_on'] / df_work['shots_total'], 0.5)

# качество паса
pass_qual = np.where(df_work['passes_total'] > 0,
                     df_work['passes_key'] / df_work['passes_total'], 0.05)

# веса по ролям
is_att = pos.isin(['FW','ST','CF','LW','RW','AM','CAM'])
is_mid = pos.isin(['CM','DM','CDM','LM','RM'])
is_def = pos.isin(['CB','RB','LB','RWB','LWB'])

goal_w = np.where(is_att, 1.0, np.where(is_mid, 0.6, 0.3))
assist_w = np.where(is_att, 0.8, np.where(is_mid, 1.0, 0.5))
pass_w = np.where(is_mid, 1.2, np.where(is_def, 0.8, 0.6))

# скор производительности
perf_score = (
    goal_w * np.log1p(g_pm * 10) +
    assist_w * np.log1p(a_pm * 10) +
    0.3 * shot_acc +
    pass_w * pass_qual * 2 +
    0.2 * np.log1p(df_work['passes_total'] / 100)
)
perf_factor = 1 + np.clip(perf_score / 3, 0, 1.5)

# активность
total_activity = (
    df_work['shots_total'] * 2 +
    df_work['goals'] * 10 +
    df_work['assists'] * 8 +
    df_work['passes_key'] * 3 +
    df_work['passes_total'] * 0.1
)
activity_factor = 1 + np.log1p(total_activity) / 15

# капитан
captain_factor = np.where(df_work['is_captain'].fillna(0).astype(bool), 1.08, 1.0)

# рост
height = df_work['player_height'].replace(0, np.nan).fillna(180)
height_z = (height - 180) / 10
height_matter = pos.isin(['CB','ST','CF','FW','GK'])
height_factor = np.where(height_matter, 1 + 0.02 * np.tanh(height_z), 1.0)

# базовая стоимость
base_value = (
    500_000 +
    total_activity * 150 +
    df_work['goals'] * 800_000 +
    df_work['assists'] * 500_000
)

# итоговая оценка
df_work['estimated_value'] = (
    base_value *
    age_factor *
    pos_factor *
    league_factor *
    perf_factor *
    activity_factor *
    captain_factor *
    height_factor
)

# глобальный масштаб: 2.0 вместо 0.90 для совпадения с Transfermarkt
df_work['estimated_value'] *= 2.0

# границы и округление
df_work['estimated_value'] = df_work['estimated_value'].clip(200_000, 200_000_000)
df_work['estimated_value'] = (df_work['estimated_value'] / 100_000).round() * 100_000
```

```python
# молодость
youth_factor = np.exp(-((df_work['age'] - 21) ** 2) / (2 * (5 ** 2)))

# нормализуем рейтинг
rating_norm = (df_work['player_rating'] - df_work['player_rating'].min()) / (df_work['player_rating'].max() - df_work['player_rating'].min())

# нормализуем цену
price_norm = (df_work['estimated_value'] - df_work['estimated_value'].min()) / (df_work['estimated_value'].max() - df_work['estimated_value'].min())
df_work['player_potential'] = rating_norm * youth_factor * (1 - price_norm)

# масштабируем 0-100
df_work['player_potential'] = (df_work['player_potential'] / df_work['player_potential'].max()) * 100
```
