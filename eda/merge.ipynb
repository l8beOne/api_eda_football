```python
import pandas as pd
contracts = pd.read_csv(f"{os.path.dirname(current_dir)}/data_collection/contracts.csv")
players = pd.read_csv(f"{os.path.dirname(current_dir)}/data_collection/players_stats.csv")
```

```python
char_replacements = {
        'æ': 'ae', 'Æ': 'AE',
        'ø': 'o', 'Ø': 'O',
        'å': 'a', 'Å': 'A',
        'ß': 'ss', 'ẞ': 'SS',
        'ñ': 'n', 'Ñ': 'N',
        'ç': 'c', 'Ç': 'C',
        'ğ': 'g', 'Ğ': 'G',
        'ş': 's', 'Ş': 'S',
        'ı': 'i', 'İ': 'I',
        'ö': 'o', 'Ö': 'O',
        'ü': 'u', 'Ü': 'U',
        'ä': 'a', 'Ä': 'A',
        'é': 'e', 'É': 'E',
        'á': 'a', 'Á': 'A',
        'í': 'i', 'Í': 'I',
        'ó': 'o', 'Ó': 'O',
        'ú': 'u', 'Ú': 'U',
        'ê': 'e', 'Ê': 'E',
        'â': 'a', 'Â': 'A',
        'î': 'i', 'Î': 'I',
        'ô': 'o', 'Ô': 'O',
        'û': 'u', 'Û': 'U',
        'ë': 'e', 'Ë': 'E',
        'ï': 'i', 'Ï': 'I',
        'ÿ': 'y', 'Ÿ': 'Y',
        'č': 'c', 'Č': 'C',
        'š': 's', 'Š': 'S',
        'ž': 'z', 'Ž': 'Z',
        'ń': 'n', 'Ń': 'N',
        'ł': 'l', 'Ł': 'L',
        'đ': 'd', 'Đ': 'D',
        'ħ': 'h', 'Ħ': 'H',
        'ć': 'c', 'Ć': 'C',
        'ę': 'e', 'Ę': 'E',
        'ą': 'a', 'Ą': 'A',
        'ś': 's', 'Ś': 'S',
        'ź': 'z', 'Ź': 'Z',
        'ż': 'z', 'Ż': 'Z'
    }
for old_char, new_char in char_replacements.items():
    players['player_name'] = players['player_name'].str.replace(old_char, new_char, regex=False)

def to_initials(df, name_column='player_name', new_column='new_name'):
    df[new_column] = df[name_column].str[0] + '.' + df[name_column].str.split().str[-1]
    return df
to_initials(players, name_column = 'player_name', new_column = 'new_name')
```

```python
for old_char, new_char in char_replacements.items():
    contracts['player'] = contracts['player'].str.replace(old_char, new_char, regex=False)

to_initials(contracts, name_column = 'player', new_column = 'new_name')
```

```python
df = pd.merge(players, contracts, on='new_name')
df
```

```python
df.to_csv('merged_df.csv', index=False)
```
